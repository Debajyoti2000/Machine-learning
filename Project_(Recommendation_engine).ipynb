```python
cols=['price','review_data','product_descrip',"price_range"]
data.loc[:,'combined']=data[cols].apply(lambda row: ' '.join(row.values.astype(str)),axis=1)
```

```python
stop_words=stopwords.words('english')
data.loc[:,'combined']=data['combined'].str.lower().apply(lambda x: x.strip().split(' '))
data.loc[:,'combined_formatted']=data['combined'].apply(lambda x:[word for word in x if word not in stop_words])
```

```python
data.tail()
```

```python
data['combined_formatted'] = data['combined_formatted'].apply(lambda x: str(x))
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(data["combined_formatted"])
```

```python
print(count_matrix)
```

```python
cosine_sim=cosine_similarity(count_matrix)
```

```python
def getTop10Similar(productName):
    index=data[data.product_name==productName].index.values.tolist()
    sim_scores=list(enumerate(cosine_sim[index[0]]))
    sorted_similar_products=sorted(sim_scores,key=lambda x: x[1],reverse=True)
    top10=sorted_similar_products[:10]
    df_n=pd.DataFrame()
    for element in top10:
        new_row=data[data.index==element[0]]
        df_n=df_n.append(new_row,ignore_index=True)
    return df_n
```
